```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("Data_set_cleaned.csv")

df.head(50)
```

```python
# analyse univarié des charges

print(df["charges"].describe())

plt.hist(df["charges"],rwidth=.5,bins=15)
plt.xlabel("Charges")
plt.ylabel("Nombre de personnes")
plt.show()
```

```python
# analyse univarié du BMI

print(df["bmi"].describe())

plt.hist(df["bmi"],rwidth=.5)
plt.xlabel("BMI")
plt.ylabel("Nombre de personnes")
plt.show()
```

```python
# analyse univarié des enfants

print(df["children"].describe())

plt.hist(df["children"],rwidth=.5, bins=6)
plt.xlabel("Enfants")
plt.ylabel("Nombre de personnes")
plt.show()
```

```python
# analyse univarié du sex

print(df["sex"].describe())

counts = df["sex"].value_counts()

labels = counts.index
sizes = counts.values
colors = ['skyblue', 'pink']
plt.pie(sizes,labels=labels, autopct='%1.1f%%',colors=colors)
plt.title("Proportion du sex")
plt.show()

```

```python
# analyse univarié des fumeurs

print(df["smoker"].describe())

counts = df["smoker"].value_counts()

labels = counts.index
sizes = counts.values
colors = ['green', 'red']
plt.pie(sizes,labels=labels, autopct='%1.1f%%',colors=colors)
plt.title("Proportion des fumeurs")
plt.show()
```

```python
# analyse univarié des enfants

print(df["region"].describe())

plt.hist(df["region"],rwidth=.25, bins=4)
plt.xlabel("region")
plt.ylabel("Nombre de personnes")
plt.show()
```

```python
# étude de la correlation avec la variable cible

df_test = pd.DataFrame({'A': [1, 2, 3, 4], 'B': [5, 6, 7, 8]})

# Define a function to transform the values
def transform(x):
    return x * 2

# Apply the transformation to the 'A' column
df_test['A'] = df_test['A'].apply(transform)

# Print the modified dataframe
print(df_test)
```

```python
df_corr = df

def transform(x):
    if x == "man" or x == "no":
        x = 0
    else:
        x = 1
    return x

df_corr["sex"] = df_corr["sex"].apply(transform)
df_corr["smoker"] = df_corr["smoker"].apply(transform)


df_corr.corr()
```

```python
from scipy.stats import ttest_ind, f_oneway


cont_var = df['charges']
cat_var = df['sex']

# T-test

t_stat, p_val = ttest_ind(cont_var[cat_var == 'Category 1'], cont_var[cat_var == 'Category 2'])
print('T-test statistic:', t_stat)
print('p-value:', p_val)

# ANOVA test
f_stat, p_val = f_oneway(cont_var[cat_var == 'Category 1'], cont_var[cat_var == 'Category 2'])
print('ANOVA test statistic:', f_stat)
print('p-value:', p_val)
```
